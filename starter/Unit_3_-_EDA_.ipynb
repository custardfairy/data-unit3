### Plot for 'age' column

```python
plt.figure(figsize=(12,6))
ax = sns.histplot(data=df, x='age', bins=8)
plt.xlabel("age", fontsize=12)
plt.ylabel("Number of Managers", fontsize=12)
plt.title("Distribution of Manager Age Range")
```

### Plot for 'industry' column

```python
plt.figure(figsize=(10,10))
data = df['industry'].value_counts()
labels = ['business', 'finance', 'technology', 'healthcare']
colors = sns.color_palette('Accent')
plt.title ("Industries in Survey", fontsize=16);
ax = plt.pie(data, colors = colors, labels = labels, autopct = '%.0f%%');
```

##### Insights:

The four industries included are approximately equal in representation, around a quarter of the data per industry.

### Plot for 'salary' column

```python
sns.displot(data = df, x = 'salary', kind = 'hist', bins=30)

plt.xlabel("Salary", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Distribution of Salary Among Managers", fontsize=16);
```

##### Insights:

The mean appears to be around the 100K mark with salaries tapering steeply under and over 100K.  This is an inconsistent bell curve.

### Plot for 'industry_experience' column

```python
plt.figure(figsize=(12,6))

ax = sns.histplot(data=df, x='industry_experience', bins=20)

plt.xlabel("Industry Experience(in years)", fontsize=14)
plt.ylabel("Number of Managers", fontsize=14)
plt.title("Industry Experience Among Managers", fontsize=16);
```

##### Insights:

The majority of managers have between 15 and 25 years of industry experience.  Experience outside of that range is lower.  There appears to be a marked drop in experience after 5, 12, and 25 years respectively with a marked increase to follow.

### Plot for 'education' column

### Step 1 - Review Summary Statistics

- mean
> the average; equal to the sum of all values in the data set divided by the number of values in the data set.
- median
> the middle score for a set of data arranged by order of magnitude.
- mode
> the most frequently occuring value in a data set; the number or value repeated most often.
- central tendency
> a central or typical value for a probability distribution, also called a measure of central location.
- distribution
> a collection of data on a variable or value; frequency refers to how often an event occurs while probability refers to the likelihood that an event will occur.
- variance
> the spread between numbers or values in a data set calculated by using the average of squared deviations from the mean.
- standard deviation
> measures the dispersion of a dataset relative to its mean and is calculated as the square root of the variance.

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = there is a positive correlation between salary and education.

```python
spearmanr(df['salary'], df['education'])
```

There is a small positive correlation between education and salary.

#### Are these results statistically significant? How do you know?

The pvalue is below .05 at 0, which is significant.

### What is the correlation between salary and age?

#### Your Hypothesis: There is a positive correlation between salary and age 

```python
pearsonr(df['salary'], df['age'])
```

There is a slight positive correlation between salary and age.

#### Are these results statistically significant? How do you know? 

The pvalue is below .05 at 0, which is significant.

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = [...]

```python
spearmanr(df['salary'], df['industry_experience'])
```

There is a slight positive correlation between salary and industry_experience.

#### Are these results statistically significant? How do you know?

The pvalue is below .05 at 0, which is significant.

```python
crosstab = pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
plt.figure(figsize=(16,10))

ax = sns.heatmap(crosstab, annot=True, linewidths=.5, cmap='YlGnBu', fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])

plt.xlabel("Education Level", fontsize=14)
plt.ylabel("Industry", fontsize=14)

plt.title("Manager Salaries Across Industries by Education Level", fontsize=18);
```

The biggest jump in salary in terms of education level is from completion of high school to completion of some college across all industries. Healthcare is the least well-paid among the industry variables and this holds across the board regardless of education level.  Technology followed by finance show the biggest salary increases dependent on level of education.
